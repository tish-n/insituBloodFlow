```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

# Define the range of x values
x = np.arange(-10, 10, 0.1)
percent_xscale = 0.5
percent_yscale = 1
xscale = 1 * percent_xscale
yscale = 1 * percent_yscale
xshift = 0
yshift = -4

# Define the bell curve using the e^(-x^2) formula
y = yscale * np.exp(-((1/xscale)*x - xshift/xscale)**2) + yshift

# Create a 3D figure
plt.style.use('dark_background')
plt.rcParams['axes.prop_cycle'] = plt.cycler(color=['#FFC61E','#8C92AC'])
plt.rcParams['axes.facecolor'] = '#BDBEC6'
plt.rcParams['axes.edgecolor'] = 'black'
plt.rcParams['axes.linewidth'] = 1


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
fig.set_dpi(250)
fig.set_linewidth(1)
fig.set_tight_layout(True)
fig.set_size_inches((10, 10))

# Plot the bell curve rotated around the x-axis
for i in range(len(x)):
    theta = np.linspace(0, 2*np.pi, 100)
    yt = y[i] * np.sin(theta)
    zt = y[i] * np.cos(theta)
    # ax.plot(x[i]*np.ones(len(theta)), yt, zt, color='red', alpha=0.5)
    ax.plot(x[i]*np.ones(len(theta)), yt, zt, alpha=0.72)

# Set the axis labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Rotated Bell Curve')

# Set the axis limits
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.set_zlim(-10, 10)


# Show the plot
plt.show()
```
